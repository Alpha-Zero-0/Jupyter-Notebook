## String

### String Interpolation
The process of inserting variables and expressions into a string

```python
name = 'Chewbacca'
species = 'wookiee'
print(f'{name} is a {species}')
```

### Translation Table (Example - Cipher Encryption)

```python
alphabet = 'abcdefghijklmnopqrstuvwxyz'
shift = 5
shifted_alphabet = alphabet[shift:] + alphabet[:shift]
translation_table = str.maketrans(alphabet, shifted_alphabet)
text = 'hello world'
encrypted_text = text.translate(translation_table)
print(encrypted_text)
```

### Common String Method

```python
ship = ' Home One '
print(1, ship.lower())
print(2, ship.upper())
print(3, ship.strip())
print(4, ship.replace('One', 'Two'))
print(5, ship.split())
arr = ship.split()
print(6, ' '.join(arr))                   #iterable element into a string with whatever it's in the ''
print(7, ship.startswith(' '))            #check if the prefix is ''
print(8, ship.endswith('One'))            #suffix
print(9, ship.find('One'))                #index of first occurrence of a substring
print(10, ship.count('O'))                 #case sensitive!
character = 'leia organa'
print(11, character.capitalize())          #ONLY capitalise the first char in a str
print(12, character.isupper())             #ALL uppercase?
print(13, character.islower())
print(14, character.title())               #.capitalize() pro max
```

## Function & Scope

### Scope - LEGB

| Scope | Definition | Location |
| :--- | :--- | :--- |
| **L**ocal | Variables defined inside a function or class. | Inside a standard `def` block. |
| **E**nclosing | Variables in the outer function of a nested pair. | In-between the outer and inner functions. |
| **G**lobal | Variables defined at the top level of the script or module. | Outside of any functions or classes. |
| **B**uilt-in | Keywords and functions pre-loaded by Python. | System defaults like `print()`, `len()`, `int`. |

## List

### Common Methods

```python
fleet1 = ['Home One', 'Rebel B-wing']
fleet2 = ['Raven Claw', 'Cassian U-wing']
fleet1.extend(fleet2)
print(fleet1)
```

```python
fleet1.insert(2, 'Ghost')
print(fleet1)
```

```python
fleet1.remove('Ghost')
print(fleet1)
```

```python
fleet1.pop(1)
print(fleet1)
```

```python
fleet1.sort()                               # Modifying the original list
print(fleet1)
fleet3 = sorted(fleet1)                     # Creating new list
print(fleet3)
```

```python
fleet1.reverse()
print(fleet1)
```

```python
fleet4 = sorted(fleet1, key=len)            # Sort by length
print(fleet4)
```

### Enumerate

```python
for index, ship in enumerate(fleet1, 1):      # index, item (order!)
    print(f'{ship} at {index}')               # optional parameter - starting index (default 0)
```

### Zip

```python
for i in zip(fleet1, fleet2):              # zip 2 lists, pair elements of the same index, return 1 tuple
    print(i)

print(list(zip(fleet1, fleet2)))           # len(zip) = min(len(list_1), len(list_2))
```

### List Comprehension

```python
numbers = [1, 2, 3, 4, 5]
result = [(num, 'Even') if num % 2 == 0 else (num, 'Odd') for num in numbers]
print(result)
```

```python
numbers = [1, 2, 3, 4, 5, 6]                                    # filter(Boolean, list)
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))      # mini lambda function + filter
print(even_numbers)                                             # lambda: function with no name
```

### Map

```python
celsius = [0, 10, 20, 30, 40]                                   # map(function, list)

def to_fahrenheit(temp):
    return (temp * 9/5) + 32

fahrenheit = list(map(to_fahrenheit, celsius))
print(fahrenheit)
```

## Dictionary

```python
fleet = {
    'name': 'Raven Claw',
    'category': ['Jedi', 'Rebel'],
    'type': 'support'
} 
print(fleet['name'])                                # dictionary[key]
print(fleet.get('type', None))                      # dictionary.get(key, default)     default is used when key doesn't exist

print(fleet.values())
print(fleet.keys())
print(fleet.items())

# fleet.clear        remove everything

print(fleet.pop("type", 10))                            # dictionary.pop(key, default)   remove a pair; key doesn't exist -> return default
print(fleet)

fleet.update({                                          # dictionary1.update(dictionary2)  add new items or overwrite dictionary1 values
    'type': 'attacker',
    'name': 'whatever'
})
print(fleet)
```

## Set

```python
squad = set()                                # Define an empty set
squad = {'Great Mothers', 'Morgan Elsbeth', 'Nightsister Spirit', 'Death Trooper', 'Night Trooper'}

squad.add('me')
print(squad)

squad.discard('me')
print(squad)

# squad.clear()                              # Remove everything

squad2 = {'Great Mothers', 'Morgan Elsbeth'}
print(squad2.issubset(squad))                # A.issubset(B) = A is subset of B?

print(squad.isdisjoint(squad2))              # A.isdisjoint(B) = no common elements?

print(squad | squad2)                        # A | B = union

print(squad & squad2)                        # A & B = intersection

print(squad - squad2)                        # A - B = difference

print(squad ^ squad2)                        # A ^ B = A or B but NOT in both   (Symmetric Difference)
```
