## String

### String Interpolation
The process of inserting variables and expressions into a string

```python
name = 'Chewbacca'
species = 'wookiee'
print(f'{name} is a {species}')
```

### Translation Table (Example - Cipher)

```python
alphabet = 'abcdefghijklmnopqrstuvwxyz'
shift = 5
shifted_alphabet = alphabet[shift:] + alphabet[:shift]
translation_table = str.maketrans(alphabet, shifted_alphabet)
text = 'hello world'
encrypted_text = text.translate(translation_table)
print(encrypted_text)
```

### Common String Method

```python
ship = ' Home One '
print(1, ship.lower())
print(2, ship.upper())
print(3, ship.strip())
print(4, ship.replace('One', 'Two'))
print(5, ship.split())
arr = ship.split()
print(6, ' '.join(arr))                   #iterable element into a string with whatever it's in the ''
print(7, ship.startswith(' '))            #check if the prefix is ''
print(8, ship.endswith('One'))            #suffix
print(9, ship.find('One'))                #index of first occurrence of a substring
print(10, ship.count('O'))                 #case sensitive!
character = 'leia organa'
print(11, character.capitalize())          #ONLY capitalise the first char in a str
print(12, character.isupper())             #ALL uppercase?
print(13, character.islower())
print(14, character.title())               #.capitalize() pro max
```

## Function & Scope

### Scope - LEGB

| Scope | Definition | Location |
| :--- | :--- | :--- |
| **L**ocal | Variables defined inside a function or class. | Inside a standard `def` block. |
| **E**nclosing | Variables in the outer function of a nested pair. | In-between the outer and inner functions. |
| **G**lobal | Variables defined at the top level of the script or module. | Outside of any functions or classes. |
| **B**uilt-in | Keywords and functions pre-loaded by Python. | System defaults like `print()`, `len()`, `int`. |

## List

### Common Methods

```python
fleet1 = ['Home One', 'Rebel B-wing']
fleet2 = ['Raven Claw', 'Cassian U-wing']
fleet1.extend(fleet2)
print(fleet1)
```

```python
fleet1.insert(2, 'Ghost')
print(fleet1)
```

```python
fleet1.remove('Ghost')
print(fleet1)
```

```python
fleet1.pop(1)
print(fleet1)
```

```python
fleet1.sort()                            # Modifying the original list
print(fleet1)
fleet3 = sorted(fleet1)                  # Creating new list
print(fleet3)
```

```python
fleet1.reverse()
print(fleet1)
```

```python
fleet4 = sorted(fleet1, key=len)       # Sort by length
print(fleet4)
```
