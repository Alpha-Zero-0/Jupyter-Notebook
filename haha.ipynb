## String

### String Interpolation
The process of inserting variables and expressions into a string

```python
name = 'Chewbacca'
species = 'wookiee'
print(f'{name} is a {species}')
```

### Common String Method

```python
ship = ' Home One '
print(ship.lower())
print(ship.upper())
print(ship.strip())
print(ship.replace('One', 'Two'))
print(ship.split())
arr = ship.split()
print(' '.join(arr))                   #iterable element into a string with whatever it's in the ''
print(ship.startswith(' '))            #check if the prefix is ''
print(ship.endswith('One'))            #suffix
print(ship.find('One'))                #index of first occurrence of a substring
print(ship.count('O'))                 #case sensitive!
character = 'leia organa'
print(character.capitalize())          #ONLY capitalise the first char in a str
print(character.isupper())             #ALL uppercase?
print(character.islower())
print(character.title())               #.capitalize() pro max
```

## Boolean

```python
print(bool(False))             # Common falsy values
print(bool(0))  
print(bool(''))
print(bool(None))
print(bool(0.0))
```

```python
print(36 and True)            # if False: return 1st operand
print(0 and 25)               # otherwise: return 2nd
print(False and ' ')
```
